```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

```python
import pandas as pd
df = pd.read_csv('/content/archive (5).zip')
```

```python
#Distribution of Employees by Department
plt.figure(figsize=(10,6))
sns.countplot(data=df,x='Department')
plt.title('Distribution of Employees by Department')
plt.xticks(rotation=45)
plt.show()
```

```python
#Salary Distribution by Department
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='Department', y='Salary')
plt.title('Salary Distribution by Department')
plt.xticks(rotation=45)
plt.show()
```

```python
# Performance Scores Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='PerfScoreID', bins=10, kde=True)
plt.title('Performance Scores Distribution')
plt.show()
```

```python
# Employee Satisfaction Levels
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='EmpSatisfaction', bins=10, kde=True)
plt.title('Employee Satisfaction Levels')
plt.show()
```

```python
# Absences vs. Performance Scores
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Absences', y='PerfScoreID')
plt.title('Absences vs. Performance Scores')
plt.show()
```

```python
# Engagement Survey Scores vs. Employee Satisfaction
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='EngagementSurvey', y='EmpSatisfaction')
plt.title('Engagement Survey Scores vs. Employee Satisfaction')
plt.show()
```

```python
# Days Late in Last 30 Days Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='DaysLateLast30', bins=10, kde=True)
plt.title('Days Late in Last 30 Days Distribution')
plt.show()
```

```python
# recruitment sources distribution
recruitment_counts = df['RecruitmentSource'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(recruitment_counts, labels=recruitment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Recruitment Source Distribution')
plt.axis('equal')
plt.show()
```

```python
# Convert 'DateofHire' to datetime and extract year
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df['Year'] = df['DateofHire'].dt.year
hires_per_year = df['Year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.plot(hires_per_year.index, hires_per_year.values, marker='o')
plt.title('Number of Hires Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Hires')
plt.grid(True)
plt.show()
```
